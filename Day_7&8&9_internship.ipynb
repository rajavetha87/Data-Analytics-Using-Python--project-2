```python
import pandas as pd
data={
    'emp_nm':['John','Mary'],
    'emp_sal':[25000,30000]
}
df=pd.DataFrame(data)
df.rename(columns={
    'emp_nm':'Employee_name',
    'emp_sal':'Employee_salary'
    },inplace=True)
print(df)
```

```python
import pandas as pd
data={
    "Region":["south","north","east","west","south","north"],
    "products":["laptops","mobiles","laptop","tablet","mobile","tablet"],
    "sales":[50000,30000,45000,25000,20000,35000]
}
df=pd.DataFrame(data)
print(df)
```

```python
south_sales=df[df["Region"]=="south"]
print(south_sales)
```

```python
high_sales=df[df["sales"]>30000]
print(high_sales)
```

```python
sorted_df=df.sort_values("sales")
print(sorted_df)
```

```python
region_sales=df.groupby("Region")["sales"].sum()
print(region_sales)
```

```python
print(df["sales"].sum())
```

```python
print(df["sales"].max())
```

```python
print(df["sales"].min())
```

```python
print(df["sales"].count())
```

```python
result=df.groupby("Region")["sales"].agg(
    ["sum","mean","max","min"]
    )
print(result)
```

```python
import matplotlib.pyplot as plt
months=["Jan","Feb","Mar","Apr","May","Jun"]
sales=[10000,12000,15000,13000,18000,20000]
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(months,sales,
         marker="o")
plt.title("sales trend")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.subplot(1,2,2)
plt.bar(months,sales,
        color="orange")
plt.title("sales comparison")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```
