```python
df=pd.read_csv("train.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df.head()
```

```python
df['Hour'] =pd.to_datetime(df['Time_Orderd'],errors='coerce').dt.hour
peak_orders =df['Hour'].value_counts().sort_index()
peak_orders.plot(kind='bar')
plt.title("Peak Orders")
plt.xlabel("Hour")
plt.ylabel("Orders")
plt.show()
```

```python
popular_dish=df['Type_of_order'].value_counts()
popular_dish.head(10).plot(kind='bar')
plt.title("Popular Dish")
plt.xlabel("Food Item")
plt.ylabel("Orders")
plt.show()
```

```python
location_orders=df['City'].value_counts()
location_orders.plot(kind='pie',autopct='%1.1f%%')
plt.title("Orders by City")
plt.ylabel('')
plt.show()
```

```python
df['Time_taken(min)'] = df['Time_taken(min)'].str.replace('(min) ', '', regex=False).astype(int)
city_delivery = df.groupby('City')['Time_taken(min)'].mean()
city_delivery.plot(kind='bar')
plt.title("Average Delivery Time by City")
plt.xlabel("City")
plt.ylabel("Delivery Time (min)")
plt.show()
```

```python
fig, ax=plt.subplots(2,2,figsize=(12,10))
peak_orders.plot(kind='bar',ax=ax[0,0])
ax[0,0].set_title("Peak Orders")
popular_dish.head(5).plot(kind='bar',ax=ax[0,1])
ax[0,1].set_title("Popular Dish")
city_delivery.plot(kind='bar',ax=ax[1,0])
ax[1,0].set_title("Average Delivery Time by City")
location_orders.plot(kind='pie',autopct='%1.1f%%',ax=ax[1,1])
ax[1,1].set_title("Orders by City")
plt.tight_layout()
plt.show()
```
