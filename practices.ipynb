```python
# Product Sales Performance Analysis

# Product sales data
products = {
    "Laptop": 50000,
    "Mobile": 35000,
    "Tablet": 20000,
    "Headphones": 10000,
    "Smartwatch": 15000
}

# Total sales
total_sales = sum(products.values())
print("Total Sales:", total_sales)

# Average sales
average_sales = total_sales / len(products)
print("Average Sales:", average_sales)

# Best selling product
best_product = max(products, key=products.get)
print("Best Selling Product:", best_product)
print("Sales Amount:", products[best_product])

# Least selling product
least_product = min(products, key=products.get)
print("Least Selling Product:", least_product)
print("Sales Amount:", products[least_product])

# Sales performance report
print("\n--- Sales Performance Report ---")
for product, sales in products.items():
    if sales > average_sales:
        status = "Above Average"
    elif sales < average_sales:
        status = "Below Average"
    else:
        status = "Average"

    print(f"{product}: ₹{sales} - {status}")
```

```python
# Employee Dataset Exploration

import pandas as pd

# Sample Employee Dataset
data = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Name": ["Arun", "Priya", "Karthik", "Divya", "Vijay"],
    "Department": ["HR", "IT", "Finance", "IT", "HR"],
    "Salary": [35000, 60000, 55000, 65000, 40000],
    "Experience": [2, 5, 4, 6, 3]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display Dataset
print("Employee Dataset:")
print(df)

# Basic Information
print("\nDataset Information:")
print(df.info())

# Statistical Summary
print("\nStatistical Summary:")
print(df.describe())

# Average Salary
avg_salary = df["Salary"].mean()
print("\nAverage Salary:", avg_salary)

# Highest Paid Employee
highest_paid = df.loc[df["Salary"].idxmax()]
print("\nHighest Paid Employee:")
print(highest_paid)

# Employees by Department
print("\nEmployees by Department:")
print(df["Department"].value_counts())

# Employees with Salary Above Average
print("\nEmployees with Salary Above Average:")
print(df[df["Salary"] > avg_salary])
```

```python
import pandas as pd
import numpy as np

# Create a dummy sales_data.csv for demonstration if it doesn't exist
try:
    with open("sales_data.csv", "x") as f:
        f.write("Product,Sales,Date\nLaptop,50000,2023-01-01\nMobile,35000,2023-01-02\nTablet,20000,2023-01-03\nHeadphones,10000,2023-01-04\nSmartwatch,15000,2023-01-05\nLaptop,,2023-01-06\nMobile,40000,invalid-date\nTablet,22000,2023-01-08\nLaptop,55000,2023-01-09\n")
except FileExistsError:
    pass # File already exists, proceed

# Load dataset
df = pd.read_csv("sales_data.csv")

print("Original Dataset:")
print(df)

# Remove duplicate rows
df = df.drop_duplicates()

# Handle missing values
df["Product"] = df["Product"].fillna("Unknown")
df["Sales"] = df["Sales"].fillna(df["Sales"].mean())

# Remove extra spaces from text columns
df["Product"] = df["Product"].str.strip()

# Standardize text format
df["Product"] = df["Product"].str.title()

# Convert date column to proper format
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Remove rows with invalid dates
df = df.dropna(subset=["Date"])

# Convert Sales column to numeric
df["Sales"] = pd.to_numeric(df["Sales"], errors="coerce")

# Remove rows with invalid sales values
df = df.dropna(subset=["Sales"])

# Reset index
df = df.reset_index(drop=True)

print("\nCleaned Dataset:")
print(df)

# Save cleaned dataset
df.to_csv("cleaned_sales_data.csv", index=False)

print("\nDataset cleaned successfully!")
```

```python
import pandas as pd

# Sample Sales Dataset
data = {
    "Region": ["North", "South", "East", "West", "North", "South"],
    "Product": ["Laptop", "Mobile", "Tablet", "Laptop", "Mobile", "Tablet"],
    "Sales": [50000, 30000, 20000, 45000, 35000, 25000]
}

# Create DataFrame
df = pd.DataFrame(data)

print("Sales Dataset:")
print(df)

# Region-wise Total Sales
region_sales = df.groupby("Region")["Sales"].sum()

print("\nRegion Wise Sales Report:")
print(region_sales)

# Highest Sales Region
highest_region = region_sales.idxmax()
highest_sales = region_sales.max()

print("\nHighest Sales Region:", highest_region)
print("Sales Amount:", highest_sales)
```
